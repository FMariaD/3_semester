```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as so
import csv


x_data, y_data = [], []
with open("Segment(max).txt", encoding='utf-8') as r_file:
    
    file = csv.reader(r_file, delimiter = ";")
    for row in file:
            x_data.append(row[1])
            y_data.append(row[0])

                      
for i in range(0, len(x_data)):
    x_data[i] = float(x_data[i])
for i in range(0, len(y_data)):
    y_data[i] = float(y_data[i])

plt.figure(figsize=(18, 10))
plt.plot(x_data, y_data, 'ro', label=r'experimental', color='orange')
plt.xlabel(r'$Max\ Number$', fontsize=14)
plt.ylabel(r'$Probability$', fontsize=14)
plt.title(r'$SegmentState$', fontsize=16)
plt.legend(loc='best', fontsize=12)
plt.grid(True)
                      
plt.show()
```
